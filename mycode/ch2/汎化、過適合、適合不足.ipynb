# 汎化、過剰適合、適合不足
- 過剰適合:学習データに特化させすぎて、未知のデータで精度が上がらない
- 適合不足:学習データに対しても未知のデータの対しても精度が上がらない  
この二つはトレードオフなので、最良の精度を示すつまり汎化性が一番よいスイートスポットを見つけることが良いモデリングだと言える  
  
データセットに含まれるデータがバリエーションに富んでいれば、より複雑で説得力のあるモデリングができる。少量のデータで得られた知見より大量のデータで得られた知見の方が説得力があるのは一般的なことだ。**実問題においてはモデルを弄くり回すより単純にデータ量を増やしたほうが良い場合が多い。データ量が持つ力を舐めてはいけない**(正則化でその効果を感じることができる)

以降のアルゴリズムで使用するサンプルデータセットの確認

```python
import mglearn
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#分類で使用する単純なデータセット
#データ数26で二つの特徴量を持つ。クラスは2つ
X,y = mglearn.datasets.make_forge()

mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.legend(['Class 0','Class 1'],loc=4)
plt.xlabel('First Feature')
plt.ylabel('Second Feature')
print('X Shape: {}'.format(X.shape))
```

```python
#回帰で使用するデータセット
X, y = mglearn.datasets.make_wave(n_samples=40)

plt.plot(X,y,'o')
plt.ylim(-3,3)
plt.xlabel('Feature')
plt.ylabel('Target')
```

```python
#より多いデータを使う場合はsklern内のデータセットを使う

#分類の場合ウィスコンシン乳がんデータセット
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print('Keys :{}'.format(cancer.keys()))
print('Features:\n{}'.format(cancer.feature_names))
print('Shape of data:{}'.format(cancer.data.shape))
print('sample count per class:\n{}'.format({n: v for n, v in zip(cancer.target_names,np.bincount(cancer.target))}))
```

```python
#回帰の場合はボストン近郊の住宅価格
from sklearn.datasets import load_boston
boston = load_boston()
print('keys:{}'.format(boston.keys()))
print('features:\n{}'.format(boston.feature_names))
print('Data shape:{}'.format(boston.data.shape))
```
